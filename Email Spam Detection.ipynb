# Spam/Ham Email Detection

```python
# Workflow of Machine Learning Model
'''
0) Import Important dependencies
1) Setting up the data
2) Creating the Model
3) Training the Model
4) Testing the Model
5) Prediction
'''
```

## 1. Import libraries

```python
# Importing Important Dependencies
import numpy as np
import pandas as pd
```

```python
# Importing LogisticRegression Model(Binary Output)
from sklearn.linear_model import LogisticRegression
```

```python
# Importing train test split function
from sklearn.model_selection import train_test_split
```

```python
# Import accuracy function
from sklearn.metrics import accuracy_score
```

## 2. Load dataset

```python
# Importing the dataset or Load dataset
df = pd.read_csv('Mail_data.csv')
df
```

## 3. Data preprocessing

```python
# Information of the dataset
df.info()
```

```python
# Checking null values
df.isnull().sum()
```

```python
# Basic Analysis
# Check the ham/spam emails counts
df['Category'].value_counts().reset_index()
```

```python
# Label Enconding
# Labelling spam email as 0 and ham emails as 1
# df.loc[df['diagnosis'] == 'B', 'diagnosis'] = 1
df.loc[df['Category'] == 'ham','Category'] = 1
df.loc[df['Category'] == 'spam','Category'] = 0
```

```python
# Seperating the Mail data into Input Data and Label(Output Data)
X = df['Message']
Y = df['Category']
```

```python
X
```

```python
Y
```

## 4. Train-test-split

```python
# Splitting the data into Training data and Testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
# Feature Extraction
# Transforming the text into numeric machine understanding data
from sklearn.feature_extraction.text import TfidfVectorizer
feature_extraction = TfidfVectorizer(min_df = 1, stop_words = 'english', lowercase = True)
```

```python
X_train_num = feature_extraction.fit_transform(X_train)
X_test_num = feature_extraction.transform(X_test)
```

```python
# Convert Y_train and Y_test values as integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

## 5. Train Random Forest model

```python
# Training the model
model = LogisticRegression(max_iter = 1000)
model.fit(X_train_num, Y_train)
```

## 6. Evaluate model accuracy

```python
# Evaluating the Trained Model
X_train_prediction = model.predict(X_train_num)
# Finding the accuracy score of training data
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print("Accuracy score of Training data is", training_data_accuracy)
```

```python
# Evaluating the testing Model
X_test_prediction = model.predict(X_test_num)
# Finding the accuracy score of testing data
testing_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print("Accuracy score of Testing data is", testing_data_accuracy)
```

## 7. Predictive system

```python
# Build a predictive system
input_data = ['Click on this link to avail free free free money']
input_data_num = feature_extraction.transform(input_data)
```

```python
# Making Prediction
prediction = model.predict(input_data_num)
print("Email Text:", input_data[0])

if prediction[0] == 1:
    print("Prediction: Ham Email")
else:
    print("Prediction: Spam Email")
```
